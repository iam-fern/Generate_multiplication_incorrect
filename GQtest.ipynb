```python
# ลบข้อมูลที่มี missing values
table = df.dropna()
table
```

```python
table.reset_index(inplace=True)
table
```

```python
table.set_index(['rep', 'index'], inplace=True)
```

```python
table
```

```python
# แสดงค่าของ Replicate ที่ 1.0
new_table = table.loc[0.0]
new_table
```

```python
# เรียง data จากน้อยไปมากตาม x1
new_table_sort = new_table.sort_values('x2')
new_table_sort
```

```python
new_table_sort.reset_index(inplace=True)
new_table_sort
```

```python
# แบ่งข้อมูลกลุ่มที่ 1
#table1 = new_table_sort.iloc[:12 , :]
#table1

table1 = new_table_sort.iloc[:int(0.4*new_table_sort.shape[0]) , :]
table1
```

```python
table2 = new_table_sort.iloc[int((1-0.4)*new_table_sort.shape[0]): , :]
table2
```

```python
# define respones variable
y = table1['Y']

# define predictor variable
x = table1[['x1', 'x2', 'x3']]

# add constant to predictor variable
x = sm.add_constant(x)

# fit linear regression model
model1 = sm.OLS(y, x).fit()

# view model summary
# print(model.summary())
```

```python
# หาค่า df กลุ่ม 1
DF1 = model1.df_resid+1
DF1
```

```python
# หาค่า SSE1 : Residual sum of squares
SSE1 = model1.ssr
SSE1
```

```python
MSE1 = SSE1/DF1
MSE1
```

```python
y = table2['Y']
x = table2[['x1', 'x2', 'x3']]
x = sm.add_constant(x)
model2 = sm.OLS(y, x).fit()
#print(model.summary())
```

```python
# หาค่า df กลุ่ม 2
DF2 = model2.df_resid+1
DF2
```

```python
# หาค่า SSE2
SSE2 = model2.ssr
SSE2
```

```python
MSE2 = SSE2/DF2
MSE2
```

```python
GQ_test = MSE2 / MSE1
GQ_test
```
